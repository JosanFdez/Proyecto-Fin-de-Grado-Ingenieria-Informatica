##### Iteramos sobre cada tipo de aeronave para poder extraer sus datos

```python
for aircraft in aircraft_types:
    url = f"{base_url}{aircraft}"
    print(f"Accediendo a {url}")
    driver.get(url)
    time.sleep(4)

    # Cargar más resultados si hay botón "Próximos 40"
    for _ in range(5):
        try:
            next_button = driver.find_element(By.LINK_TEXT, "Próximos 40")
            next_button.click()
            time.sleep(3)
        except:
            break

    # Extraer filas de la tabla
    rows = driver.find_elements(By.CSS_SELECTOR, "table.prettyTable tbody tr")
    for row in rows:
        cols = row.find_elements(By.TAG_NAME, "td")
        if len(cols) >= 6:
            record = {
                "Ident": cols[0].text,
                "Tipo": aircraft,
                "Origen": cols[2].text,
                "Destino": cols[3].text,
                "Salida": cols[4].text,
                "Llegada estimada": cols[5].text
            }
            all_data.append(record)

driver.quit()
```

#### Procedo a arreglar un poco los datos para obtener los datos desagregados por columnas y calcular la duración

```python
import pandas as pd

# Cargar el archivo
archivo = "vuelos_por_tipo_aeronave_2.xlsx"  # Asegúrate de que esté en el mismo directorio
df = pd.read_excel(archivo)

# Función para dividir fecha, hora y zona horaria
def dividir_fecha_hora_zona(col):
    fechas, horas, zonas = [], [], []
    for val in df[col]:
        if pd.isna(val):
            fechas.append(None)
            horas.append(None)
            zonas.append(None)
            continue
        partes = str(val).split()
        if len(partes) == 3:
            fechas.append(partes[0])
            horas.append(partes[1])
            zonas.append(partes[2])
        else:
            fechas.append(None)
            horas.append(None)
            zonas.append(None)
    return fechas, horas, zonas

# Aplicar a Salida
df["Fecha salida"], df["Hora salida"], df["Zona horaria salida"] = dividir_fecha_hora_zona("Salida")

# Aplicar a Llegada estimada
df["Fecha llegada"], df["Hora llegada"], df["Zona horaria llegada"] = dividir_fecha_hora_zona("Llegada estimada")

# Guardar resultado si lo deseas
df.to_excel("vuelos_con_fechas_separadas_2.xlsx", index=False)

# Mostrar primeras filas como verificación
print(df[["Ident", "Tipo", "Salida", "Llegada estimada", 
          "Fecha salida", "Hora salida", "Zona horaria salida", 
          "Fecha llegada", "Hora llegada", "Zona horaria llegada"]].head())
```

```python
import pandas as pd
from datetime import datetime
import pytz

# Cargar el archivo original
df = pd.read_excel("vuelos_por_tipo_aeronave_2.xlsx")

# Diccionario de zonas horarias comunes mapeadas a pytz
timezone_map = {
    "UTC": "UTC", "CST": "Asia/Shanghai", "BST": "Europe/London", "WIB": "Asia/Jakarta",
    "+08": "Asia/Singapore", "+07": "Asia/Bangkok", "AEST": "Australia/Sydney", "KST": "Asia/Seoul",
    "EDT": "America/New_York", "CDT": "America/Chicago", "MDT": "America/Denver", "PDT": "America/Los_Angeles",
    "AST": "America/Puerto_Rico", "EST": "America/New_York", "CEST": "Europe/Madrid", "IST": "Asia/Kolkata"
}

# Funciones para separar y convertir

def separar_fecha_hora_zona(col):
    fechas, horas, zonas = [], [], []
    for val in df[col]:
        if pd.isna(val):
            fechas.append(None)
            horas.append(None)
            zonas.append(None)
            continue
        partes = val.strip().split()
        if len(partes) == 3:
            fechas.append(partes[0])
            horas.append(partes[1])
            zonas.append(partes[2])
        else:
            fechas.append(None)
            horas.append(None)
            zonas.append(None)
    return fechas, horas, zonas

def convertir_a_datetime(fecha, hora, zona):
    if pd.isna(fecha) or pd.isna(hora) or pd.isna(zona):
        return pd.NaT
    try:
        dt_naive = datetime.strptime(f"{fecha} {hora}", "%a %I:%M%p")
        tz_name = timezone_map.get(zona.strip(), "UTC")
        tz = pytz.timezone(tz_name)
        return tz.localize(dt_naive)
    except Exception:
        return pd.NaT

# Separar columnas
(df["Fecha salida"], df["Hora salida"], df["Zona horaria salida"]) = separar_fecha_hora_zona("Salida")
(df["Fecha llegada"], df["Hora llegada"], df["Zona horaria llegada"]) = separar_fecha_hora_zona("Llegada estimada")

# Convertir a datetime
df["Salida datetime"] = df.apply(lambda row: convertir_a_datetime(row["Fecha salida"], row["Hora salida"], row["Zona horaria salida"]), axis=1)
df["Llegada datetime"] = df.apply(lambda row: convertir_a_datetime(row["Fecha llegada"], row["Hora llegada"], row["Zona horaria llegada"]), axis=1)

# Calcular duración
df["Duración (min)"] = (df["Llegada datetime"] - df["Salida datetime"]).dt.total_seconds() / 60

# Eliminar negativos
df.loc[df["Duración (min)"] < 0, "Duración (min)"] = None

# Formato hh:mm
df["Duración (hh:mm)"] = df["Duración (min)"].apply(lambda x: f"{int(x // 60):02}:{int(x % 60):02}" if pd.notna(x) else None)

# Guardar
output_file = "vuelos_con_duracion_final_2.xlsx"
df.to_excel(output_file, index=False)
print(f"Archivo guardado como {output_file}")

```
