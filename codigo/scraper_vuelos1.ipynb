#### Inicializa lista de resultados

```python
excel_path = "3.vuelos_resultado.xlsx"
all_flights = []

if os.path.exists(excel_path):
    all_flights = pd.read_excel(excel_path).to_dict(orient="records")

total_airports = len(airlines)

for idx, code in enumerate(airlines, start=1):
    print(f"[{idx}/{total_airports}] Procesando aeropuerto: {code}")
    url = base_url + code
    driver.get(url)
    time.sleep(5)

    # Inicializar vuelos del aeropuerto
    airport_flights = []

    # Recoger vuelos en la página inicial
    for i in range(5):  # 5 bloques de 40 = 200 vuelos
        if i > 0:
            try:
                offset = i * 40
                next_url = f"{base_url}{code}?;offset={offset};order=ident;sort=ASC"
                driver.get(next_url)
                time.sleep(4)
            except Exception as e:
                print(f"Error al intentar acceder al offset {offset}: {e}")
                break

        rows = driver.find_elements(By.CSS_SELECTOR, "table.prettyTable tbody tr")
        for row in rows:
            cols = row.find_elements(By.TAG_NAME, "td")
            if len(cols) >= 6:
                flight = {
                    "Ident": cols[0].text,
                    "Tipo": cols[1].text,
                    "Origen": cols[2].text,
                    "Destino": cols[3].text,
                    "Salida": cols[4].text,
                    "Llegada estimada": cols[5].text,
                    "Aerolínea": code
                }
                airport_flights.append(flight)

        # Break anticipado si ya hay 200 vuelos por cualquier motivo
        if len(airport_flights) >= 200:
            break

    # Limitar a 200 en caso de que haya más
    airport_flights = airport_flights[:200]

    # Acumular y guardar
    all_flights.extend(airport_flights)
    df = pd.DataFrame(all_flights)
    df.to_excel(excel_path, index=False)
    print(f"Aeropuerto {code} completado con {len(airport_flights)} vuelos. Faltan {total_airports - idx} aeropuertos.\n")

driver.quit()
print("Todos los aeropuertos procesados y guardados.")


```

#### Para retocar el excel, aplico lo siguiente

```python
import pandas as pd

# Cargar el archivo original
df = pd.read_excel("3_vuelos_todas_aerolineas.xlsx")

# Función corregida para dividir fecha, hora, y zona horaria
def dividir_fecha_hora_zona(col):
    fechas, horas, zonas = [], [], []
    for val in df[col]:
        if isinstance(val, str):
            partes = val.split()
            if len(partes) == 3:
                fechas.append(partes[0])
                horas.append(partes[1])
                zonas.append(partes[2])
            else:
                fechas.append(None)
                horas.append(None)
                zonas.append(None)
        else:
            fechas.append(None)
            horas.append(None)
            zonas.append(None)
    return fechas, horas, zonas

# Aplicar a "Salida" y "Llegada estimada"
df["Fecha salida"], df["Hora salida"], df["Zona horaria salida"] = dividir_fecha_hora_zona("Salida")
df["Fecha llegada"], df["Hora llegada"], df["Zona horaria llegada"] = dividir_fecha_hora_zona("Llegada estimada")

# Mostrar ejemplo
print(df[["Ident", "Fecha salida", "Hora salida", "Zona horaria salida",
          "Fecha llegada", "Hora llegada", "Zona horaria llegada"]].head())

# Guardar nuevo Excel con columnas divididas
df.to_excel("3_vuelos_con_fechas_separadas.xlsx", index=False)
```

#### Calculo de la duración de los vuelos, teniendo en cuenta las zonas horarias. 

```python
from datetime import datetime, timedelta
import pandas as pd

df = pd.read_excel("3_vuelos_con_fechas_separadas.xlsx")

# Función para combinar fecha y hora en texto y convertir a datetime
def combinar_datetime(fecha, hora):
    try:
        return datetime.strptime(f"{fecha} {hora}", "%a %I:%M%p")
    except:
        return pd.NaT  # Not a Time

# Aplicar a cada fila
df["Salida datetime"] = df.apply(lambda row: combinar_datetime(row["Fecha salida"], row["Hora salida"]), axis=1)
df["Llegada datetime"] = df.apply(lambda row: combinar_datetime(row["Fecha llegada"], row["Hora llegada"]), axis=1)

# Convertir columnas a tipo datetime explícitamente (por si acaso)
df["Salida datetime"] = pd.to_datetime(df["Salida datetime"], errors="coerce")
df["Llegada datetime"] = pd.to_datetime(df["Llegada datetime"], errors="coerce")

# Calcular duración en minutos
df["Duración (min)"] = (df["Llegada datetime"] - df["Salida datetime"]).dt.total_seconds() / 60

# Calcular duración como hh:mm
def minutos_a_hhmm(minutos):
    try:
        if pd.isna(minutos):
            return None
        td = timedelta(minutes=minutos)
        total_seconds = int(td.total_seconds())
        hours, remainder = divmod(total_seconds, 3600)
        minutes = remainder // 60
        return f"{hours:02d}:{minutes:02d}"
    except:
        return None

df["Duración (hh:mm)"] = df["Duración (min)"].apply(minutos_a_hhmm)

# Mostrar resultados
print(df[["Ident", "Salida datetime", "Llegada datetime", "Duración (min)", "Duración (hh:mm)"]].head())

# Guardar resultado en nuevo archivo
df.to_excel("3_vuelos_con_duracion.xlsx", index=False)



```
